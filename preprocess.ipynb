```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
data_path = './Data'
data_dir_list = os.listdir(data_path)
print(data_dir_list)

save_path='./PreImage'

img_data_list=[]

face_haar_cascade=cv2.CascadeClassifier('Cascade Classifier/haarcascade_frontalface_default.xml')

for dataset in data_dir_list:
    img_list=os.listdir(data_path+'/'+ dataset)
    print ('Loaded the images of dataset-'+'{}\n'.format(dataset))
    print (len(img_list))
    for img in img_list:
        input_img=cv2.imread(data_path + '/'+ dataset + '/'+ img )
        roi_gray = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
        faces_detected=face_haar_cascade.detectMultiScale(roi_gray,scaleFactor=1.3,minNeighbors=5)
        for(x,y,w,h) in faces_detected:
            cv2.rectangle(roi_gray,(x,y),(x+w,y+h),(0,255,0),thickness=0)
            crop_image=roi_gray[y:y+w,x:x+h]
            cv2.imwrite(str(w) + str(h) + '_faces.jpg', crop_image)
        
        img_data_list.append(crop_image)
        plt.imshow(crop_image)
        plt.title(str(dataset))
        plt.show()
```
